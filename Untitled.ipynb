```python
import numpy as np
from scipy.spatial import Delaunay

def calculate_euler_characteristic(k=30, tolerance=1e-6):
    ''' Counts and prints the number of vertices, edges, faces (triangles), 
        and calculates the Euler characteristic of the triangulated sphere.

    Parameters: 
        k: determines how fine the triangulation should be

    Returns:
        None
    '''

    U = np.linspace(0, 2 * np.pi, k)
    V = np.linspace(0, np.pi, k)
    [U, V] = np.meshgrid(U, V)

    S1 = np.cos(U) * np.sin(V)
    S2 = np.sin(U) * np.sin(V)
    S3 = np.cos(V)

    tri = Delaunay(np.array([S1.flatten(), S2.flatten(), S3.flatten()]).T)

    # Count vertices
    unique_vertices = np.unique(tri.points, axis=0)
    total_vertices = len(unique_vertices)

    # Count edges
    unique_edges = set()
    for triangle in tri.simplices:
        for i in range(3):
            edge = tuple(sorted([triangle[i], triangle[(i+1) % 3]]))
            unique_edges.add(edge)
    total_edges = len(unique_edges)

    # Count faces (triangles)
    total_faces = len(tri.simplices)

    # Calculate Euler characteristic
    euler_characteristic = total_vertices - total_edges + total_faces
    
     # Check for close vertices
    close_vertex_indices = []
    for i in range(len(tri.points)):
        for j in range(i + 1, len(tri.points)):
            distance = np.linalg.norm(tri.points[i] - tri.points[j])
            if distance < tolerance:
                close_vertex_indices.append((i, j))
                
    
    # Print indices and coordinates of close vertices
    if close_vertex_indices:
        print("Close Vertex Indices:")
        for i, j in close_vertex_indices:
            print(f"Vertices {i} and {j}:")
            print(tri.points[i])
            print(tri.points[j])
    else:
        print("No close vertices found.")
        
        
     # Find duplicate vertices and keep only one instance of each unique vertex
    unique_vertex_indices = []
    unique_vertices = []
    for i in range(len(tri.points)):
        is_duplicate = False
        for j in range(len(unique_vertices)):
            distance = np.linalg.norm(tri.points[i] - unique_vertices[j])
            if distance < tolerance:
                is_duplicate = True
                break
        if not is_duplicate:
            unique_vertex_indices.append(i)
            unique_vertices.append(tri.points[i])

    # Update the triangulation to use the unique vertices
    updated_triangles = []
    for simplex in tri.simplices:
        updated_simplex = [unique_vertex_indices.index(vertex) for vertex in simplex]
        updated_triangles.append(updated_simplex)
        
    # Check AGAIN for close vertices
    close_vertex_indices = []
    for i in range(len(updated_triangles.points)):
        for j in range(i + 1, len(updated_triangles.points)):
            distance = np.linalg.norm(updated_triangles.points[i] - updated_triangles.points[j])
            if distance < tolerance:
                close_vertex_indices.append((i, j))
                
    
    # Print indices and coordinates of close vertices
    if close_vertex_indices:
        print("Close Vertex Indices:")
        for i, j in close_vertex_indices:
            print(f"Vertices {i} and {j}:")
            print(updated_triangles.points[i])
            print(updated_triangles.points[j])
    else:
        print("No close vertices found.")
        
    # Find duplicate vertices again and keep only one instance of each unique vertex
    for i in range(len(updated_triangles.points)):
        is_duplicate = False
        for j in range(len(unique_vertices)):
            distance = np.linalg.norm(tri.points[i] - unique_vertices[j])
            if distance < tolerance:
                is_duplicate = True
                break
        if not is_duplicate:
            unique_vertex_indices.append(i)
            unique_vertices.append(tri.points[i])
        

    # Print counts and Euler characteristic
    print("Total Number of Vertices:", total_vertices)
    print("Total Number of Edges:", total_edges)
    print("Total Number of Faces (Triangles):", total_faces)
    print("Euler Characteristic:", euler_characteristic)

# Usage:
k = 10  # You can adjust k as needed
calculate_euler_characteristic(k)
```

```python
import numpy as np
from scipy.spatial import Delaunay

def calculate_euler_characteristic_and_remove_duplicates(k=30, tolerance=1e-6):
    ''' Calculates the Euler characteristic, number of edges, and removes duplicate vertices 
        from the triangulated sphere.

    Parameters: 
        k: determines how fine the triangulation should be
        tolerance: threshold for considering vertices as duplicates (adjust as needed)

    Returns:
        Euler characteristic, number of edges, unique vertices, and the updated triangulation
    '''

    U = np.linspace(0, 2 * np.pi, k)
    V = np.linspace(0, np.pi, k)
    [U, V] = np.meshgrid(U, V)

    S1 = np.cos(U) * np.sin(V)
    S2 = np.sin(U) * np.sin(V)
    S3 = np.cos(V)

    tri = Delaunay(np.array([S1.flatten(), S2.flatten(), S3.flatten()]).T)

    # Find duplicate vertices and keep only one instance of each unique vertex
    unique_vertex_indices = []
    unique_vertices = []
    for i in range(len(tri.points)):
        is_duplicate = False
        for j in range(len(unique_vertices)):
            distance = np.linalg.norm(tri.points[i] - unique_vertices[j])
            if distance < tolerance:
                is_duplicate = True
                break
        if not is_duplicate:
            unique_vertex_indices.append(i)
            unique_vertices.append(tri.points[i])

    # Update the triangulation to use the unique vertices
    updated_triangles = []
    for simplex in tri.simplices:
        updated_simplex = [unique_vertex_indices.index(vertex) for vertex in simplex if vertex in unique_vertex_indices]
        if len(updated_simplex) == 3:
            updated_triangles.append(updated_simplex)

    # Calculate Euler characteristic
    euler_characteristic = len(unique_vertices) - len(updated_triangles) + len(tri.simplices)

    # Calculate the number of edges
    num_edges = len(updated_triangles) * 3 // 2  # Each triangle contributes 3 edges, but counted twice

    return euler_characteristic, num_edges, unique_vertices, updated_triangles

# Usage:
k = 10  # You can adjust k as needed
euler_char, num_edges, unique_vertices, updated_triangles = calculate_euler_characteristic_and_remove_duplicates(k)
print("Euler Characteristic:", euler_char)
print("Number of Edges:", num_edges)
print("Number of unique vertices:", len(unique_vertices))
```
